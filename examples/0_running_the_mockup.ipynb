# Running the UR3e Mockup

This notebook provides instructions on how to run and interact with the UR3e Mockup. **Make sure to have the environment set up as described in the [Environment Setup](0_environment_setup.ipynb) notebook before proceeding.**

## Starting the mockup

The UR3e mockup is located in the `ur3e_mockup` folder. To start the mockup, open a terminal in the **root folder** of the project and run the following command:

```bash
python -m startup.start_ur3e_mockup
```
It might take a few seconds/minutes for the mockup to start the first time, as it needs to initialize all the components.

> Note: If you encounter any issues related to RabbitMQ connection, ensure that the RabbitMQ server is running and accessible with the credentials specified in the `startup/startup.conf` file. If you created the RabbitMQ server with different credentials, please update the `username` and `password` fields in the configuration file (`startup/startup.conf`) accordingly.
