# Configuring the mockup

In this notebook, we will adjust show how the UR3e mockup can be configured via the `startup.conf` file.

The `startup.conf` file is located in the `startup` folder of the UR3e mockup project. It contains various settings that control the behavior of the mockup. When starting the mockup, it reads this configuration file to set up its parameters.

The relevant section for configuring the physical twin mockup is under `physical_twin -> pt_mockup`:

```plaintext
physical_twin: {
    pt_mockup: {
        initial_q         = [0.0, -1.57, 1.57, -1.57, -1.57, 0.0]
        speedup           = 5
        publish_frequency = 20
        max_velocity      = 60  # degrees per second
        acceleration      = 80  # degrees per second squared
    }
}
```

Here are the parameters you can adjust:
- `initial_q`: This parameter sets the initial joint positions of the robot arm in radians. You can modify these values to change the starting pose of the mockup.
- `speedup`: This parameter controls the speed multiplier for the mockup's movements. Increasing this value will make the mockup respond faster to control messages.
- `publish_frequency`: This parameter defines how often (in Hz) the mockup publishes its state messages. A higher frequency means more frequent updates.
- `max_velocity`: This parameter sets the maximum velocity (in degrees per second) for the robot arm joints. Adjusting this value will affect how quickly the mockup can move.
- `acceleration`: This parameter sets the maximum acceleration (in degrees per second squared) for the robot arm joints. This affects how quickly the mockup can change its speed.

If no velocity or acceleration is specified in the control messages sent to the mockup, these default values from the configuration file will be used.
